### 15. Show Corelation between all features of dataset using heatmap

```python
wdata=ds[["Temp_C", "Dew Point Temp_C", "Rel Hum_%",
       "Wind Speed_km/h", "Visibility_km", "Press_kPa"]].corr()
```

```python
fig=figure(figsize=(5,4))
heatmap(wdata,annot=True,linewidth=1)
show()
```
